```python
import pandas as pd
import numpy as np
import snscrape.modules.twitter as snt
username=input()
to=input()
start=input()

def tweetss(userame,to,start):

    query="(from:{username}) until:{to} since:{start}".format(username=username,to=to,start=start)
    tweets=[]
    no_tweet=100
    for tweet in snt.TwitterSearchScraper(query).get_items():

        if len(tweets)==no_tweet:
            break
        tweets.append([tweet.date,tweet.id,tweet.content,tweet.user.username])
    df=pd.DataFrame(tweets,columns=['date','id','content','username'])    

    df.to_csv('tweet.csv')


tweetss(username,to,start)

#data preprocessing
#segmenter using the word statistics from Twitter
```
